```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

file = "dataset_tcc_churn_0-1.csv" # dataset file
ds_path = f"../datasets/telco-customer-churn/versions/1/{file}" # dataset path

datas = pd.read_csv(ds_path) # header=1 eliminates header line
print(datas)
```

```python
def rename():
    for col in datas.head(0).columns:
        datas.rename(
            columns = {
                f"{col}":f"{col[0].lower()+col[1:]}"
            },
            inplace=True
        )
rename()
```
